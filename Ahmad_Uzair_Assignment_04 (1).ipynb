Create a shallow copy of a list and modify one of the lists to observe the effects on the other.


```python
listO=[[2,4],[6,8],[10,12]]

shallow_copy=listO.copy()

shallow_copy[0][0]='changed'

print("listO:", listO)
print("Shallow Copy:", shallow_copy)
```

Create a deep copy of a nested list and modify one of the lists to observe the effects on the other.

```python
import copy
listO=[[2,4],[6,8],[10,12]]

deep_copy=copy.deepcopy(listO)

deep_copy[0][0]='changed'

print("listO:", listO)
print("Deep Copy:", deep_copy)
```

Pickle and Unpickle a list and display its content.

```python
import pickle

my_list=[1,2,3,4,5]
with open ("my list.pkl",'wb')as file:
    pickle.dump(my_list,file)
    
print("list has been pickled")


with open ("my-list.pkl",'rb')as file:
    loaded_list=pickle.load(file)

print("unoickled List", loaded_list)
```
